```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#importing the data into a pandas dataframe
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```python
print(concrete_data.describe())
print(concrete_data.isnull().sum())
```

```python
#Assigning the label variable
y = concrete_data.Strength

#Assigning the independent variables i.e the predictors
X = concrete_data[concrete_data.columns[concrete_data.columns != 'Strength']]
```

# A - Baseline model

## Step 1: Spliting the data into train and test data of 30%

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
#Checking the shapes of the split data
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
n_cols = X_train.shape[1]
```

```python
#Defining my regression model
def Regression_Model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu',input_shape = (n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
```

```python
#Build model
model = Regression_Model()
```

## Step 2: Training the model on the training data 

```python
#Fit model
model.fit(X_train, y_train, epochs = 50, verbose = 2)
```

## Step 3: Evaluating the model on the test data to compute the mean squared error between the predicted concrete strength and the actual concrete strength

```python
y_pred = model.predict(X_test)

MSE = mean_squared_error(y_test,y_pred)
MSE
```

## Step 4. Repeating steps 1 - 3 50 times

```python
#Putting it all in a loop to run 50 times to get 50 mean squared errors
errors = []

for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
    model.fit(X_train, y_train, epochs = 50, verbose = 2)
    y_pred = model.predict(X_test)
    errors.append(mean_squared_error(y_test,y_pred))
errors
```

# Step 5: Fetching the mean and standard deviation of the 50 mean squared errors

```python
mean_errors = np.array(errors).mean()
std_errors = np.array(errors).std()
print(mean_errors,std_errors)
```
